```python
from token_predictor import TokenPredictor
```

```python
timesteps = [3, 10, 50, 100, 150, 200]
```

```python
predictions = []

for timestep in timesteps:
    tp = TokenPredictor()
    tp.timesteps = timestep
    tp.train()
    predictions.append(tp.predict())
```
